```python
# create a new column for total flow for each row || run once
df['TotalFlow'] = df['Flow(Lane 1)'] + df['Flow(Lane 2)'] + df['Flow(Lane 3)'] + df['Flow(Lane 4)'] + df['Flow(Lane 5)'] + df['Flow(Lane 6)'] + df['Flow(Lane 7)']
df['datetime'] = pd.to_datetime((df['Date'] + " " + df['Time']), infer_datetime_format=True)
df['hour'] = df['datetime'].dt.hour
df['date'] = df['datetime'].dt.date
#df['Time'] = pd.to_datetime(df['Time'],format= '%H:%M' )
```

```python
M1_total_hr = df.groupby(["Geographic Address", "date", "hour"])["TotalFlow"].sum()
M1_total_hr
```
